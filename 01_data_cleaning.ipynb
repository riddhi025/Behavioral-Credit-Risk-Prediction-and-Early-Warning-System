####Project Overview
######This project aims to detect early-warning signals of customer financial stress using historical credit card behavior. Instead of focusing solely on default prediction, the goal is to identify deteriorating behavioral patterns before default occurs, enabling proactive intervention by financial institutions.

Business Problem
Credit card issuers face significant risk when customers silently drift into financial stress. Traditional default models are reactive and identify risk too late. This project focuses on identifying leading indicators of financial stress using customer payment behavior over time.

### Raw Data Source

The dataset is sourced from the UCI Machine Learning Repository and
contains customer demographics, billing history, payment behavior,
and default outcomes.


```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00350/default%20of%20credit%20card%20clients.xls"
raw_df = pd.read_excel(url, header=None)
```

```python
raw_df.columns = raw_df.iloc[1]
df = raw_df.iloc[2:].copy()

df.columns = df.columns.astype(str).str.replace('\n', ' ').str.strip()
df.rename(columns={'default payment next month': 'default'}, inplace=True)

df.reset_index(drop=True, inplace=True)
df = df.apply(pd.to_numeric, errors='raise')
```

### Data Cleaning Summary

Column names were standardized, data types were converted to numeric,
and the target variable was renamed for clarity.

Data Source
The dataset is sourced from the UCI Machine Learning Repository and originates from a real-world credit card default study (Yeh & Lien, 2009), published in Expert Systems with Applications. It contains demographic and behavioral data for credit card clients, collected over a six-month period.


####Temporal Structure of Customer Behavior
###### Although the dataset is not a traditional time series, it contains ordered monthly behavioral snapshots for each customer (e.g., payment status, bill amounts, and payment amounts over six months).

###### This longitudinal structure allows the modeling of behavioral trends, deterioration, recovery, and volatility, which are critical for detecting early signs of financial stress.
